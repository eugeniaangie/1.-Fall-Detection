```python
import os
sleeping_dir=os.path.join('./Dataset/sleeping')
sitting_dir=os.path.join('./Dataset/Sitting')
standing_dir=os.path.join('./Dataset/Standing')
print('Total training sleeping images:', len(os.listdir(sleeping_dir)))
print('Total training sitting images:', len(os.listdir(sitting_dir)))
print('Total training standing images:', len(os.listdir(standing_dir)))
sleeping_files=os.listdir(sleeping_dir)
sitting_files=os.listdir(sitting_dir)
standing_files=os.listdir(standing_dir)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
pic_index = 2

next_sleeping=[os.path.join(sleeping_dir, fname) for fname in sleeping_files[pic_index-2:pic_index]]
next_sitting=[os.path.join(sitting_dir, fname) for fname in sitting_files[pic_index-2:pic_index]]
next_standing=[os.path.join(standing_dir, fname) for fname in standing_files[pic_index-2:pic_index]]

#show the pict in the folder
for i, img_path in enumerate(next_sitting+next_sleeping+next_standing):
    img=mpimg.imread(img_path)
    plt.imshow(img)
    plt.axis('Off')
    plt.show()
```
